### Merge DataFrames

```python
india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python

us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
df = pd.concat([india_weather, us_weather])
df
# simple concat attaches one df below other. we see the index values are also attached.
```

```python
# to get continual index vals, we need to set ignore_index=True
df = pd.concat([india_weather, us_weather], ignore_index=True)
df

# find equivalent sql
```

```python
# it is possible to give keys to refer the parent db. 
df = pd.concat([india_weather, us_weather], keys=["india", "us"])
```

#### Merging data horizontally, along columns
we need to include axis = 1 parameter

```python
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
}, index=[0,1,2])

temperature_df
# in this df dehli has index 1
```

```python


windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "windspeed": [7,12],
}, index=[1,0])
windspeed_df

# therefore here also delhi should have index 1, so we have index = [1,0]
```

```python
# now we merge data horizontally
df = pd.concat([temperature_df,windspeed_df],axis=1)
df

# blr is not there in second df, so NaN
```

```python
# merging df and series:
s = pd.Series(["Humid","Dry","Rain"], name="event")
s
```

```python
# merging df and series:
df = pd.concat([temperature_df,s],axis=1)
df
```

```python
# Merging , similar to SQL joins for two tables

df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

#### In Pivot function, we can have one of the columns as x-axis- row(index parameter) and as y-axis(column parameter)

```python
df.pivot(index='city', columns='day')
# here on row index, we will have city, and we will get temp, windspeed, event for each date
```

#### Crosstab- For contingency table or frequency distribution

```python
df = pd.read_excel("pandas_revision.xlsx", sheet_name="crosstab_data")
```

```python
df
```

```python
pd.crosstab(df.Nationality,df.Handedness)
# here natitonality comes rowwise or as x-axis, and handedness values : left and right in column (y-axis)
```

```python
# one more example
pd.crosstab(df.Sex,df.Handedness)
```

```python
pd.crosstab(df.Sex,df.Handedness, margins=True)
```

```python
pd.crosstab(df.Sex, [df.Handedness,df.Nationality], margins=True)

# here x-axis or row data will be male and female, in y-axis/columns we have handedness as first header and nationality as second
```

```python
pd.crosstab([df.Nationality, df.Sex], df.Handedness, margins=True)

# here row-wise  we have two level nationality first, sex vals for each nation 
```

```python
pd.crosstab(df.Sex, df.Handedness, normalize='index')
# instead of giving count, it gives percentage.
```

```python
pd.crosstab(df.Sex, df.Handedness, values=df.Age, aggfunc=np.average)
```
