```python
country_grp = df.groupby(['Country'])
country_grp.get_group('India')



# in the group by dataframe : i..e : country_grp, we can use the countries as index for loc fn
# country_grp['column_name'].median().loc('country_name')

# It is because, in pandas groupby fn has a parameter: as_index, it makes the column that we use for grouping as the index, by
# default it is true, thus you can use it directly in loc

# country_grp.loc['Country'] 
# refer: https://stackoverflow.com/questions/41236370/what-is-as-index-in-groupby-in-pandas/41237258

# first you grp the entire dataframe based on the country column
# secong you get only those rows based on group selected here India
# this is equivalen to filtering

# filt = df['Country'] == 'India'
# df.loc[filt]
# to get social media counts for a single country
# df.loc[filt]['SocialMedia'].value_counts()
```

```python
filt1 = df['Country'] == 'India'
df.loc[filt1]
# gives all rows where Country = India
df.loc[filt1]['SocialMedia'].value_counts()
```

```python
#df.loc[filt]['SocialMedia'].value_counts()
#equvivalent in group by
country_grp['SocialMedia'].value_counts().loc['India']

# country_grp['Social Media'].value_counts() will give all countries social Media data
```

```python
country_grp['ConvertedComp'].median()
# will give median salary for every country
```

```python
# in the previous command, country names come first and hance are like index. so can be used in loc function to 
#fetch a particular country value
country_grp['ConvertedComp'].median().loc['Germany']
```

```python
country_grp['ConvertedComp'].median().loc[['Germany','India']]
# more than one country can come with loc fn, but u need it in a list.
```

```python
# in agg function you could pass on various statisticaal data
country_grp['ConvertedComp'].agg(['median','mean','count','min']).loc['India']

# check how to obtain percentile values
# median = 50% so, we can calculate 25 and 75% guess
```

```python
country_grp['ConvertedComp'].agg(['sum','mean','count','min'])
# can also use np, (numpy)
#df1['Item_Weight'].agg([np.sum, np.mean,np.std])
```

```python
country_grp['LanguageWorkedWith'].str.contains('Python', na= False).loc['India']
# You cannot use str function directly on group by object
# You need to use apply function
```
