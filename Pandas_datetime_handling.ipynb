```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/sathya-explr/pandas-code-challenges-3017222/main/data/stock_data.csv"
```

```python
df = pd.read_csv(url,header=0)
```

```python
df.head()
```

```python
df.info()
```

We see that the unamed column contains the date values. 
- First rename it
- covert datatype
- extract year using dt.year
- fourth set date column as index

```python
# rename the column Unnamed: 0 to Date
df.rename(columns={'Unnamed: 0':'Date'},inplace=True)
```

```python
df.info()
```

```python
# now we need to convert its datatype into datetime 
df['Date'] = pd.to_datetime(df['Date'], format = '%Y-%m-%d')
```

```python
df['Date'].dtype 
```

The result implies that we hae converte it datetime datatype

```python
df['year'] = df['Date'].dt.year
```

```python
df.head()
```

```python
# now we set index the Date column as index
df.set_index('Date', inplace=True)
```

```python
ret = df.pct_change()
# here the first row is always NaN as there is no previuos row
# it gives daily percent change
# Eg: in row index 1, 7.45-7.40/7.40 gives 0.006757
# by default the percentage change values are not multiplied by 100
```

```python
ret
```

```python
# now for the year column we do not want percentage change, we replace it with df['year']
ret['year'] = df['year']
```

```python
ret.head()
```

```python
# this is daily percentage change , for yearly percentage change , we need to groupby year column
# first we dropna and then groupby
ret.dropna()
```

```python
# yearly percentage change
ret.groupby(['year']).sum()
# ret.groupby(['year']).sum()*100 gives percentage terms. 
```
